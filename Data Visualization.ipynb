# Data Visualization

```python
import pandas as pd
```

```python
fmri = pd.read_csv("fmri.csv")
```

```python
fmri.head()
```

## Lineplot

```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
sns.lineplot(x="timepoint", y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal",hue="event",data=fmri)
plt.show()
#Grouping data with hue
```

```python
sns.lineplot(x="timepoint", y="signal",hue="event",style="event",data=fmri)
plt.show()
#Adding styles
```

```python
sns.lineplot(x="timepoint", y="signal",hue="event",style="event",markers=True,data=fmri)
plt.show()
#Adding markers
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="region")
plt.show()
```

## Barplot

```python
import pandas as pd
```

```python
sns.set(style="whitegrid")
```

```python
attentions = pd.read_csv("attentions.csv")
```

```python
sns.barplot(x = "solutions",y = "score", data = attentions)
plt.show()
```

```python
sns.barplot(x = "solutions",y = "score",hue = "subject",data = attentions)
plt.show()
```

```python
sns.barplot(x = "solutions",y = "score", data = attentions,palette = "Blues_d")
plt.show()
#using different color palettes in changing colours of the bars
```

```python
sns.barplot(x = "solutions",y = "score", data = attentions,palette = "rocket")
plt.show()
```

```python
sns.barplot(x = "solutions",y = "score", data = attentions,palette = "vlag")
plt.show()
```

```python
sns.barplot(x = "solutions",y = "score", data = attentions,color = "yellow")
plt.show()
#using a specific color that you want your bars to have
```
